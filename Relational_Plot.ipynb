```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
sns.set(style='darkgrid')
tips=sns.load_dataset("tips")
tips.head()
```

**Relational plots are used for visualizing the statistical relationship between the data points. Visualization is necessary because it allows the human to see trends and patterns in the data. The process of understanding how the variables in the dataset relate each other and their relationships are termed as Statistical analysis.**

```python
sns.relplot(x='total_bill',y='tip',data=tips)
plt.show()
```

**In seaborn, the hue parameter represents which column in the data frame, you want to use for color encoding.Adding `hue="smoker" represent that you want to use “smoker” for color encoding i.e. color the data points for smoker and non-smoker differently.**

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='size',size='size',sizes=(15,200))
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker')
sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
plt.show()
```
